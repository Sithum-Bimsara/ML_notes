#  Training and Testing in Machine Learning

##  Understanding the Data

We use **features** like:
-  Bollinger bands
-  Momentum
-  Price changes

These **features (X)** help us **predict prices (Y)**. If we have multiple features, we represent them as:
- X1, X2, X3, 

##  Splitting the Data
To scientifically evaluate our **machine learning algorithms**, we split the data into:
-  **Training Data** (Xtrain, Ytrain)
-  **Testing Data** (Xtest, Ytest)

 If we trained and tested on the same data, the results would be **unreliable**. 
 The technique of separating them is called **out-of-sample testing**  a crucial step! 

##  The Machine Learning Process
1 Take **Xtrain** and **Ytrain** 
2 Feed them into a **machine learning algorithm** (like Linear Regression or KNN) 
3 Generate a **model** 
4 Test it using **Xtest** 
5 Compare the predicted Y with the actual Ytest 

##  Time-Oriented Data
 As you move **downward**, you're moving **forward in time**.
 We **train on older data** and **test on newer data**.
 **Training on later data and testing on earlier data** is discouraged due to **look-ahead bias**! 

 The goal: The more closely the models **predicted Y** matches **Ytest**, the **more accurate** the model is! 




###  In machine learning,
 - the typical approach is to train on old (or historical) data and test on new (or unseen) data. This allows us to evaluate how well the model generalizes to data it hasn't seen during training.

###  Reason for this approach:
**Generalization:**  The goal is to build a model that performs well not just on the data it was trained on, but also on unseen data that it will encounter in the real world. By testing on new data, you simulate how the model will behave in a real-world setting where it faces data it hasn't already learned from.

**Avoiding Overfitting:** If you were to train on new data and test on old data, the model might memorize the old data and overfit to it. This would result in poor performance on new, unseen data since the model hasnt learned how to generalize.

So, training on past data and testing on future data mimics how you would deploy a model and ensure it handles new, unseen inputs effectively.
